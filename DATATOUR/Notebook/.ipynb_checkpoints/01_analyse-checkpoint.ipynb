#### Filtrage des données par Pays 

les Pays d'afrique uniquement

```python
pays_afrique = {
    "Benin","Burkina Faso","Cote d'Ivoire","Ghana","Mali", "Niger","Nigeria","Senegal","Togo",
}

col = 'incidence_malaria_per_100'

df_afr = (
    malaria_incidence[
        (malaria_incidence['Entity'].isin(pays_afrique)) &
        (malaria_incidence['Year'].between(2000, 2022))
    ]
    .sort_values(['Entity', 'Year'])
    .reset_index(drop=True)
)

print("Lignes après filtre 14 pays + période 2000-2022:", df_afr.shape)
print("Pays effectivement présents dans le fichier paludisme:", df_afr.Entity.nunique(), "-", len(pays_afrique))

manquants = pays_afrique - set(df_afr.Entity.unique())
print("Pays africains sans aucune donnée paludisme (0 ligne 2000-2022):", sorted(manquants))
print()

couverture = df_afr.groupby('Entity').Year.count().sort_values()
print("=== Couverture temporelle incomplète (< 23 ans sur 2000-2022) ===")
incomplet = couverture[couverture < 23]
print(incomplet if len(incomplet) else "Aucun - tous les pays présents ont les 23 années")
print()
print(f"Pays avec couverture complète: {(couverture==23).sum()} / {len(couverture)}")
```

```python
### sauvegardes des donnée

import pathlib

# récupère le dossier où se trouve le notebook (DATATOUR/Notebook)
notebook_dir = pathlib.Path.cwd()

# On définit le chemin vers le nouveau dossier (DATATOUR/DATA/Data_Processed)
processed_data_dir = notebook_dir.parent / "Data_Processed"

# On crée le dossier s'il n'existe pas (parents=True évite les erreurs, exist_ok=True ne fait rien s'il existe déjà)
processed_data_dir.mkdir(parents=True, exist_ok=True)

# On sauvegarde le fichier CSV à cet endroit
df_afr.to_csv(processed_data_dir / "malariaIncidenceAfrique2000_2022.csv", index=False)

print(f"Fichier sauvegardé avec succès dans : {processed_data_dir}")
```

### EDA 

```python
df = pd.read_csv("../Data_Processed/malariaIncidenceAfrique2000_2022.csv")
df.head(10)
```

```python
col = 'Incidence of malaria (per 1,000 population at risk)'

# Mapping sous-régions
sous_region = {
    # Afrique de l'Ouest
    'Benin':'Ouest','Burkina Faso':'Ouest',"Cote d'Ivoire":'Ouest','Ghana':'Ouest',
    'Mali':'Ouest','Niger':'Ouest','Nigeria':'Ouest','Senegal':'Ouest','Togo':'Ouest',
}

df['sous_region'] = df['Entity'].map(sous_region)
non_mappes = df[df['sous_region'].isna()]['Entity'].unique()
print("Pays non mappés (à vérifier):", non_mappes)
print()

# 1) Classement global 2022 - top 10 les plus touchés
print("=== TOP 20 pays les plus touchés en 2022 ===")
top2022 = df[df.Year==2022].sort_values(col, ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(
    data = top2022,
    x = col,
    y="Entity",
    palette="Reds_r",
    hue = "Entity",
    legend = False
)

plt.title('Top 10 des pays les plus touches en 2022 par le paludisme', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Incidence Malaria sur 1000 habitants', fontsize=12)
plt.ylabel('Les Pays d\'Afrique', fontsize=12)

print("=== EVOLUTION PAR SOUS REGION ===")

df_evol = df[df['Year'].isin([2000, 2022]) & df['sous_region'].notna()]

df_region_mean = df_evol.groupby(['Year', 'sous_region'])[col].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=df_region_mean, 
    x='sous_region', 
    y=col, 
    hue='Year', 
    palette=['#4c72b0', '#c44e52']
)
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3, fontsize=10)

plt.title("Évolution de l'incidence moyenne du paludisme par sous-région (2000 vs 2022)", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Sous-région", fontsize=12)
plt.ylabel("Incidence moyenne (pour 1 000)", fontsize=12)
plt.legend(title="Année")
plt.tight_layout()
plt.show()
```

### Petite documentation

#### 1. Qualité des données brutes

2851 lignes, 116 entités, 2000–2024, aucun doublon, aucune valeur manquante (NA) au niveau brut.
Contient des lignes agrégées (World, Low-income countries, etc.) à exclure du filtre pays.

#### 2. Traitement appliqué

Filtre sur 54 pays cibles : Sénégal, Guinée, Mali, Burkina Faso, Côte d'Ivoire,Bénin etc...
Restriction à la période 2000–2022 (cadre du projet).
Résultat : 48 pays sur 53 ont des données ; 5 pays n'ont aucune donnée sur ce fichier : Lesotho, Libye, Maurice, Seychelles, Tunisie. Ce sont des pays où le paludisme n'est pas endémique (Afrique du Nord et petites îles), donc logiquement absents des indicateurs OMS de risque paludéen.

#### 3. Insights clés (EDA)
Insight clé : l'Afrique de l'Ouest reste la sous-région la plus touchée en valeur absolue (233,9 en 2022, encore au-dessus de la moyenne centrale et est-africaine), malgré une baisse comparable en pourcentage aux autres régions. Ça renforce la pertinence du choix du projet de se concentrer sur cette zone.

#### 4. Zones/périodes critiques identifiées

Le Bénin se distingue nettement : incidence bien plus fort que ses voisins sur toute la période, et la baisse la plus spectaculaire (−83,5 %), avec un décrochage net après 2008 (probablement lié à l'intensification des campagnes de lutte antipaludique — hypothèse à vérifier via une source externe, ne pas l'affirmer sans confirmation).
Mali est le pays le moins amélioré relativement : plateau élevé entre 350–440 quasi toute la période, avec seulement −12,3 % sur 22 ans — le pays le plus "à surveiller" dans votre dashboard.
Burkina Faso reste le pays avec l'incidence la plus haute en 2022 malgré une forte baisse depuis son pic 2001.

#### 5. Limites à documenter

L'indicateur est un taux d'incidence "pour 1000 habitants à risque" , il ne dit rien du nombre absolu de cas ni de la mortalité associée (colonne "décès liés au paludisme" mentionnée dans le cahier des charges mais absente de ce fichier — à signaler).
Aucune désagrégation régionale à l'intérieur d'un pays (impossible de faire la carte choroplèthe du Sénégal avec ce seul fichier — confirmera le besoin d'Africa CDC/DHIS2 mentionné dans vos sources complémentaires).

```python
df_maternal.duplicated().sum()
```

```python
df_maternal.Code.isna().sum()
```

```python
col = 'Maternal mortality ratio'

df_maternal = (
    df_maternal[
        (df_maternal['Entity'].isin(pays_afrique)) &
        (df_maternal['Year'].between(2000, 2022))
    ]
    .sort_values(['Entity', 'Year'])
    .reset_index(drop=True)
)

print("Lignes après filtre 14 pays + période 2000-2022:", df_maternal.shape)
print("Pays effectivement présents dans le fichier maternal mortality", df_maternal.Entity.nunique(), "/ 14")

manquants = pays_afrique - set(df_maternal.Entity.unique())
print("Pays africains sans aucune donnée sur la mortalité maternelle(0 ligne 2000-2022):", sorted(manquants))
print()

couverture = df_maternal.groupby('Entity').Year.count().sort_values()
print("=== Couverture temporelle incomplète (< 23 ans sur 2000-2022) ===")
incomplet = couverture[couverture < 23]
print(incomplet if len(incomplet) else "Aucun - tous les pays présents ont les 23 années")
print()
print(f"Pays avec couverture complète: {(couverture==23).sum()} / {len(couverture)}")
```

```python
df_maternal.tail(25)
```

La couverture temporelle des données sur la mortalité maternelle s'arrête 2000–2020), donc 2021 et 2022 manquent systématiquement pour la mortalité maternelle, quel que soit le pays. Donc les données supplémentaires sur Kaggle ont permis de completer cette couverture temporelle : [kaggle mortalty maternal](https://www.kaggle.com/datasets/willianoliveiragibin/maternal-mortality-ratio)

```python
data_maternal_supp = pd.read_csv("../Data_Supplementaire/maternal-mortality.csv")
data_maternal_supp.head()
```

###### On va essayer de filter ses dernieres sur les pays d'afrique cibles une merge ou melt

```python
df_maternal_b = data_maternal_supp[
    (data_maternal_supp['Entity'].isin(pays_afrique)) &
    (data_maternal_supp['Year'].between(2000, 2022))
].sort_values(['Entity', 'Year']).reset_index(drop=True)
```

```python
df_maternal_b.Year.value_counts()
```

```python
### Extraire le code de la data df_maternal

dict_codes = (
    df_maternal[['Entity', 'Code']]
    .drop_duplicates('Entity')
    .set_index('Entity')['Code']
    .to_dict()
)
df_maternal_b['Code'] = df_maternal_b['Entity'].map(dict_codes)
# concatener maintenant

df_mortality = pd.concat([df_maternal, df_maternal_b], ignore_index=True)

# trier par pays et annnée pour s'assurer de la couverture complete
df_mortality = df_mortality.sort_values(by = ['Entity', "Year"])

# suppression des doublons
df_mortality = df_mortality.drop_duplicates(subset=['Entity', 'Year'], keep='last').reset_index(drop=True)

df_mortality.head()
```

```python
### sauvegarder les doonnées

df_mortality.to_csv(processed_data_dir / "Mortality_Maternal.csv", index=False)
```

## Notes

### 1. Qualité brute

990 lignes, 180 entités, aucun doublon, aucune valeur manquante au niveau brut.
Confirmation du point déjà identifié : c'est l'indicateur avec la couverture temporelle la plus irrégulière des trois — chaque pays a des mesures ponctuelles (souvent issues d'enquêtes démographiques type EDS/MICS plutôt que d'un système de reporting annuel continu), pas une série annuelle.

### 2. Traitement appliqué

Filtre 9 pays + 2000–2022 → 64 lignes, de 5 à 12 points par pays (Sénégal le mieux couvert avec 12 points, Côte d'Ivoire le moins avec 5).
Aucune interpolation appliquée à ce stade 

### 3. Insights clés

Classement sur la dernière donnée disponible (à utiliser pour le dashboard) : Sénégal (97,6 %, 2019) > Côte d'Ivoire (93,2 %, 2016) > Guinée (80,9 %, 2018) > Burkina Faso (80,2 %, 2018) > Mali (79,5 %, 2018).
Insight fort : le Mali progresse le plus vite en relatif (parti de très bas, 56,8 %) mais reste le pays avec la couverture la plus faible en valeur absolue en dernière mesure. La Guinée est le seul pays où l'indicateur recule sur la période — point à signaler comme zone de vigilance.
### 4. Limites majeures à documenter (importantes pour la note méthodologie)

Les années de mesure diffèrent d'un pays à l'autre (ex. Sénégal 2019 vs Côte d'Ivoire 2016) → toute comparaison inter-pays "à une année donnée" est en réalité une comparaison de dernières valeurs connues, pas simultanées. À afficher clairement dans le dashboard (ex. "dernière valeur connue : [année]" à côté de chaque chiffre).
Avec seulement 5 à 12 points sur 23 ans, une courbe interpolée donnerait une fausse impression de continuité — recommandation : afficher des points/marqueurs discrets reliés par une ligne pointillée plutôt qu'une ligne pleine, pour signaler visuellement le caractère lacunaire des données.
